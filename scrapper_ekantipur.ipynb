```python
import re
# Define a regex pattern to detect Nepali text
nepali_pattern = re.compile(r'[\u0900-\u097F]')
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Base URL format
base_url = "https://ekantipur.com/News-Folder/"
url= "https://ekantipur.com/"


# File to save the articles
output_file = "scraped_ekantipur_articles.txt"
with open(output_file, "a", encoding="utf-8") as file: 
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        soup=soup.find('div', class_='nav-bar')
        soup.find('ul')
         # Extract all href attributes from <a> tags
        links = [a['href'] for a in soup.find_all('a', href=True)]
        print("Extracted Links:")
        for link in links:
            if link != "/":  # Skip the root link
                # print(link)
                link2=f"{link}{'/2025/01/18/'}"
                response = requests.get(link2)
                if response.status_code == 200:
                    soup = BeautifulSoup(response.text, 'html.parser')
                    # soup=soup.find('article', class_='normal')
                    teasers=soup.find_all('div',class_="teaser offset")
                    # Extract all href attributes from <a> tags
                    links3 = []
                    for teaser in teasers:
                        first_a_tag = teaser.find('a', href=True)
                        article_link=f"{'https://ekantipur.com'}{first_a_tag['href']}"
                        # print(first_a_tag['href'])
                        response = requests.get(article_link)
                        if response.status_code == 200:
                            soup_article= BeautifulSoup(response.text, 'html.parser')
                            # print(soup_article)
                            soup_article_heading=soup_article.find('h1', class_='eng-text eng-text-heading')

                            if soup_article_heading:
                                    soup_article_heading = soup_article_heading.get_text(strip=True)
                                    # Write article heading to file
                                    file.write("\n\n" + soup_article_heading + "\n\n")
                                    
                                    soup_paragraph = soup_article.find('div', class_='col-xs-12 col-sm-12 col-md-12')
                                    paragraphs = soup_paragraph.find_all('p')
                                    
                                    # Filter and write Nepali paragraphs to file
                                    for paragraph in paragraphs:
                                        text = paragraph.get_text(strip=True)
                                        if nepali_pattern.search(text):  # Check if Nepali characters are present
                                            file.write(text + "\n")
                                            print(text)
                            else:
                                print(f"Heading not found for article: {article_link}")
                            

                    # print(links3)    
            #new_url='https://ekantipur.com/news'
            

#response=requests.get(new_url)

```
