```python
import requests
from bs4 import BeautifulSoup

# Base URL format
base_url = "https://www.onlinekhabar.com/2025/01/"

# Article number range 
start_article_number = 1603900
end_article_number = 1604000

# File to save the articles
output_file = "scraped_onlinekhabar_articles.txt"

# Open file to write the scraped content
with open(output_file, "w", encoding="utf-8") as file:
    for article_number in range(start_article_number, end_article_number + 1):
        url = f"{base_url}{article_number}"
        response = requests.get(url)
        response.encoding = 'utf-8'

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract the headline (if exists)
            headline = soup.find('h1').get_text(strip=True) if soup.find('h1') else "No headline found"

            # Extract the date (if available)
            date_tag = soup.find('div', class_='ok-single-meta')  
            date = date_tag.get_text(strip=True) if date_tag else "No date found"

            # Extract the content (all paragraphs)
            paragraphs = soup.find_all('p')
            content_text = "\n".join([p.get_text(strip=True) for p in paragraphs])

            # Write the article to the file
            file.write(f"Article Number: {article_number}\n")
            file.write(f"URL: {url}\n")
            file.write(f"Date: {date}\n")
            file.write(f"Headline: {headline}\n\n")
            file.write(f"Content:\n{content_text}\n")
            file.write("=" * 80 + "\n\n")  # Separator for readability

            print(f"Article {article_number} saved.")
        else:
            file.write(f"Article {article_number} not found (status: {response.status_code}).\n")
            file.write("=" * 80 + "\n\n")
            print(f"Article {article_number} not found (status: {response.status_code}).")

print(f"Scraped articles have been exported to '{output_file}'.")

```
